```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],errors='coerce')
df['Ship Date']=pd.to_datetime(df['Ship Date'],errors='coerce')
df['Order year']=df['Order Date'].dt.year
```

```python
df.info()
```

```python
df['Order year'].value_counts()
```

```python
ax=sns.countplot(x='Order year',data=df,facecolor='orange')
plt.title('Sales in the calender year')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df.isnull().sum()
```

```python
df['Postal Code'].fillna(df['Postal Code'].mean(),inplace=True)
```

```python
df=df.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
df.value_counts('Ship Mode')
```

# Ship Mode

```python
plt.figure(figsize=(6,4),facecolor='orange')
ax=sns.countplot(x='Ship Mode',data=df,facecolor='orange')
plt.title('Sales on Ship mode')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df.value_counts('Segment')
```

# Customer Segment

```python
ax=sns.countplot(x='Segment',data=df)
plt.title('Sales on the basis on Segment')
ax.bar_label(ax.containers[0])
plt.show()
```

# Multivariate Analysis

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Ship Mode',data=df,hue='Category')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[1])
plt.show
```

# Region wise Sales

```python
ax=sns.countplot(x='Region',data=df)
ax.bar_label(ax.containers[0])
plt.title('Sales on the basis of Region')
plt.show()
```
