```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
d1=sns.load_dataset('tips')
```

```python
d1
```

```python
lab_enc=LabelEncoder()
```

```python
d2=lab_enc.fit_transform(d1['sex'])
```

```python
d1['sex']=d2
```

```python
d1
```

```python
d2=lab_enc.fit_transform(d1['smoker'])
d1['smoker']=d2
d2=lab_enc.fit_transform(d1['time'])
d1['time']=d2
d2=lab_enc.fit_transform(d1['day'])
d1['day']=d2
```

```python
d1
```

```python
x = d1.drop('total_bill',axis = 1)
```

```python
y = d1[['total_bill']]
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.3 , random_state = 42)
```

```python
ln = LinearRegression()
```

```python
ln.fit(x_scaled,y)
```

```python
y_predict = ln.predict(x_train)
y_predict
```

```python
x_predict = ln.predict(x_test)
x_predict
```

```python
#overFitting and underfitting
```

```python
r2_score(y_train,y_predict)
```

```python
r2_score(y_test,x_predict)
```

```python
mean_absolute_error(y_test,x_predict)
```

```python
mean_squared_error(y_test,x_predict)
```
